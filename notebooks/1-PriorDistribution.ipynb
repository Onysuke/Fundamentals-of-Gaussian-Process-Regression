```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

Gaussian Kernel $k(x,y) = \exp(-\gamma||x-y||^2)$.

```python
class GaussianKernel():
    """
    The Gaussian kernel is defined as $k(x,y) = \exp(-\gamma||x-y||^2)$.
    """
    def __init__(self, gamma):
        self.gamma = gamma
    
    def forward(self, x, y):
        return np.exp(- self.gamma * (np.abs(x - y) ** 2))
```

```python
X = np.linspace(0, 10, 100)
```

```python
gauss_kernel = GaussianKernel(gamma=1)
cov = np.array([[gauss_kernel.forward(x1, x2) for x1 in X] for x2 in X])
```

```python
mu = np.array([0] * len(X))
```

```python
plt.plot(X, np.random.multivariate_normal(mean=mu, cov=cov, size=6).T)
plt.title('Random functions generated from the prior distribution of Gaussian processes')
```

Intersection Kernel $k(x,y) = min(x, y)$.

```python
X = np.linspace(0, 10, 100)
```

```python
cov = np.array([[min(x1, x2) for x1 in X] for x2 in X])
```

```python
mu = np.array([0] * len(X))
```

```python
plt.plot(X, np.random.multivariate_normal(mean=mu, cov=cov, size=6).T)
plt.title('Random functions generated from the prior distribution of Gaussian processes')
```
